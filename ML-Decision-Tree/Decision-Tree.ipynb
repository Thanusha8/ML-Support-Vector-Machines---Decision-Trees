```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor,plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report,mean_squared_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:\\Thanusha\\Python\\ML master class\\Tree (CART)\\BreastCancer.csv")
data.head()
```

```python
x = data.iloc[:,:9].values
y = data.iloc[:,9].values
```

```python
x_tain,x_test,y_train,y_test = train_test_split (x,y, test_size=0.2 , random_state=0)
```

```python
## Classification Tree
```

```python
clt = DecisionTreeClassifier()
clt.fit(x_tain,y_train)
```

```python
plt.figure(figsize=(20,20))
plot_tree(clt, feature_names=list(data.columns), class_names=["1","2"], filled=True)
plt.show()
```

```python
y_pred = clt.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
## Using Pruning to Overcome Overfitting
```

```python
cltp = DecisionTreeClassifier(max_depth=3)
cltp.fit(x_tain,y_train)
```

```python
plt.figure(figsize=(20,20))
plot_tree(cltp, feature_names=list(data.columns), class_names=["1","2"], filled=True)
plt.show()
```

```python
y_pred = cltp.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
data = pd.read_csv("C:\\Thanusha\\Python\\ML master class\\Tree (CART)\\Boston.csv")
data.head()
```
